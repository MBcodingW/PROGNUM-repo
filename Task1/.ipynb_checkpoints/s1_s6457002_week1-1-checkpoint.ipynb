### Q1.4

```python
import math
from scipy.constants import h
frequency = 2.42e28
E = h*frequency
print(f'Energy of photon in joule: {E}')
from scipy.constants import e
E = E/e
print(f'Energy of photon in electronvolts: {E}')
```

### Q1.5 (session quiz)

```python
import math
from scipy.constants import G
M_earth = 5.9722e24
M_moon = 7.342e22
r = 385000.6e3
F = G*M_earth*M_moon/r**2
print(f'The gravitational force is {F}')
```
