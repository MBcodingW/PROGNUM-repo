### Q4.4 (session quiz)

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

lengths = np.random.normal(185, 8, 100000)

mean = np.mean(lengths)
std = np.std(lengths)

prob_less_than_avg = norm.cdf(185, 185, std)
prob_within_1_std = norm.cdf(185 + std, 185, std) - \
                    norm.cdf(185 - std, 185, std)

plt.hist(lengths, bins=40, color='skyblue', alpha=1, label=f'Data (Mean={mean:.2f}, Std={std:.2f})')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean: {mean:.2f}')
plt.axvspan(mean - std, mean + std, color='yellow', alpha=0.3, label='1 Standard Deviation Range')

plt.title('Height Distribution for 100,000 People', fontsize=15)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(fontsize='x-small')
plt.grid()
plt.show()

print(f"Calculated Mean: {mean:.4f} cm")
print(f"Calculated Std Dev: {std:.4f} cm")
print(f"P(X <= {185} cm): {prob_less_than_avg:.4f}")
print(f"P(within 1 Std Dev): {prob_within_1_std:.4f}")
```

```python
import numpy as np
from matplotlib import pyplot as plt
import math

angle_deg = 30
theta = math.radians(angle_deg)
rotor = math.cos(theta) + 1j*math.sin(theta)

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])
z = x + 1j*y
z_rot = rotor*z
xr = z_rot.real
yr = z_rot.imag

plt.plot(x, y, color='blue', marker='o')
plt.plot(xr, yr, color='red', marker='o')
plt.title('Squares')
plt.show()
```

```python
from astropy.cosmology import Planck18
from scipy import integrate
from math import sqrt

H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0
H0_Gyr = Planck18.H0.to('1/Gyr').value

#defining the derivative of t_L:
def t_Lder(z):
    Hz = H0_Gyr*sqrt(Om0*(1+z)**3+Ode0) 
    y = 1/((1+z)*Hz)
    return y

a = 0
z = 2
t_L = integrate.quad(t_Lder, a, z)
print(f'Cosmological Lookback Time Calculation gives: {t_L}.')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data matching the labels in your image: N(0,1) and N(2,1)
X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2, 1, 100000)

plt.figure(figsize=(8, 5))

# 1. Histograms with black edges
plt.hist(X, bins=30, density=True, color='blue', alpha=0.4, label='X', edgecolor='black')
plt.hist(Y, bins=30, density=True, color='red', alpha=0.4, label='Y', edgecolor='black')

# 2. PDF Lines (Thick dashed and thick solid black)
x_range = np.linspace(-6, 8, 1000)
plt.plot(x_range, norm.pdf(x_range, 0, 1), color='black', lw=4, ls='--', label='X(PDF)')
plt.plot(x_range, norm.pdf(x_range, 2, 1), color='black', lw=4, ls='--', label='Y(PDF)')

# 3. Annotations (Arrows and N labels)
plt.annotate(r'$\mathcal{N}(0,1)$', xy=(0, 0.15), xytext=(-5, 0.15),
             arrowprops=dict(facecolor='blue', shrink=0.05), color='blue', fontsize=12)
plt.annotate(r'$\mathcal{N}(2,1)$', xy=(2, 0.15), xytext=(7, 0.15),
             arrowprops=dict(facecolor='red', shrink=0.05), color='red', fontsize=12, ha='right')

# 4. Styling: Grey dashed grid and limits
plt.grid(True, axis='y', ls='--', color='grey', alpha=0.7)
plt.xlim(-6, 8)
plt.ylim(0, 0.42)
plt.xticks(np.arange(-6, 9, 1))
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.title("1D Histogram + PDF")
plt.legend()
plt.show()

plt.figure(figsize=(8, 6))

# 1. 2D Histogram with 'Blues' colormap
counts, xedges, yedges, image = plt.hist2d(X, Y, bins=30, cmap='Blues')
plt.colorbar()

# 2. Calculate centers for jagged contours
x_centers = 0.5 * (xedges[1:] + xedges[:-1])
y_centers = 0.5 * (yedges[1:] + yedges[:-1])
X_grid, Y_grid = np.meshgrid(x_centers, y_centers)

# 3. Levels calculated from the actual data to create the "indents"
# 1, 2, and 3 sigma equivalents for binned data
max_val = np.max(counts)
levels = [max_val * 0.05, max_val * 0.35, max_val * 0.65]

# 4. Draw Red Jagged Contours
contours = plt.contour(X_grid, Y_grid, counts.T, levels=levels, colors='red', linewidths=1.5)

# Label placing (1sigma, 2sigma, 3sigma)
fmt = {levels[0]: '3σ', levels[1]: '2σ', levels[2]: '1σ'}
plt.clabel(contours, inline=True, fontsize=12, fmt=fmt)

# 5. Styling: Black dashed grid
plt.grid(True, ls='--', color='black', alpha=0.8)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2D Histogram + Contour")
plt.show()
```
