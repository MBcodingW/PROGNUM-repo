### Q2.12

```python
#Angle calculation
x = float(input("Enter a value for x: "))
y = float(input("Enter a value for y: "))
from math import atan2, degrees
Angle = degrees(atan2(x, y))
print(f"The angle between the x-axis and the line between the origin and your chosen point(x,y) is {Angle} degrees.")

#Quadrant determination
if x >= 0 and y >= 0:
    print(f"The point you chose is located in the quadrant I.")
elif x < 0 and y >= 0:
    print(f"The point you chose is located in the quadrant II.")
elif x < 0 and y < 0:
    print(f"The point you chose is located in the quadrant III.")
else: print(f"The point you chose is located in the quadrant IV.")
```

The difference between the atan() and the atan2() function is that the atan() function only outputs an angle between -90 and 90 degrees, while the atan2() function can output an angle over an 360 degrees range.

### Q2.19 (session quiz)

```python
#Generating a NunPy array of ranom numbers
import numpy as np

mu = 0
sigma = 1.4
s = np.random.normal(mu, sigma, size=100000)
```

```python
#Making a graph of this array
import matplotlib.pyplot as plt

count, bins, ignored = plt.hist(s, 30, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *

               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),

         linewidth=2, color='r')

plt.show()
```

```python
#Calculating the standard deviation of the array
sd = np.std(s)
print(f"Standar deviation: {sd}.")
```
