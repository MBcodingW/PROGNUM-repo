```python
terms = range(102) #100th term in qestion = 101th term actual sequence
sum = 0
list = [0] 
for i in terms:
    print(f"Term {i} in the Fibonnaci sequence is {sum}.")
    if sum == 0:
        sum += 1
        list.append(sum)
    else:
        sum += list[-2]          #Adding up the previous sum to the current sum
        list.append(sum)
print(f"Thus the 100th term of the fibonacci sequence, as defined in the question is: {list[-2]}.")
```
