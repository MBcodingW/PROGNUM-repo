### Q3.2 (session quiz)

```python
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

escape_v = {}
from math import sqrt
from scipy.constants import G

solars = zip(names, masses, radii)
for obj in solars:
    name, M, r = obj
    v_esc = sqrt(2*G*M/r)
    print(f"Mass = {M:.2e} kg")
    escape_v[name] = v_esc

print()
print(escape_v) 
```

### Q3.4 (coursework)

```python
profile_data = [2.3, 3.5, 5.6, 12.8, 10.3, 7.2, 5.0, 2.1, -2.0]

T_sum = 0
N = 0
#Defining T_sum to be the total value of the positive value of the list.
#N counts the total amount of positive values in the list.
for data in profile_data:
    if data > 0:
        T_sum += data
        N += 1

#Calculating mean value:
mu = 1/N*T_sum
print(f"The mean value is: {mu:.2f}.")

#Calculating the total sum of all the data values plus the mean value, squared:
T_sum_xmu = 0
for data in profile_data:
    if data > 0:
        T_sum_xmu +=(data-mu)**2

#Calculating standard deviation:
from math import sqrt
sigma = sqrt(T_sum_xmu/N)
print(f"The standar deviation is: {sigma}.")
```

```python
#Importing NumPy and transforming the list into a NumPy array:
import numpy as np
numpy_data = np.array(profile_data)

#Filtering the array to only have positive values:
positive_data = numpy_data[numpy_data > 0]

#Formula for standard deviation
std = np.std(positive_data)

print(f"Using NumPy, we calculated the standard deviation to be: {std}.")
```

### Q3.5 (session quiz)

```python
total_sum = 0

#Total sum for 100, 500 and 1000 iterations:
for i in range(1, 1001):
    total_sum += 1/i**2
    if i == 100:
        x = total_sum
    elif i == 500:
        y = total_sum
    elif i == 1000:
        z = total_sum

print(f"x = {x}, y = {y}, z = {z}")

#Calculating wheter the difference between iterations gets smaller:
if (x- y) < (y -z):
    print("The function 1/x\u00b2 converges.")
else:
    print("The function does not convert.")
```

```python
from math import atan, degrees

def angular_size(d, D):
    delta = degrees(2*atan(d/(2*D)))
    return delta

print(f"d=1, D = 2 gives the angular size =", angular_size(1, 2), "degrees.")
print(f"d=7, D = 5 gives the angular size =", angular_size(7, 5), "degrees.")
```

```python
from math import pi

def cylinder(r, h):
    A = 2*pi*r**2 + 2*pi*r*h
    V = pi*r**2*h
    return A, V

area, volume = cylinder(5,10)
ratio = area/volume
print(f"The are of a spehere with radius 5 and height 10 is {area} and its volume is {volume}.")
print(f"The area-volume ratio is: {ratio}.")
```

### Q3.17 (session quiz)

```python
import numpy as np
from math import sqrt

def stdev(data):
    array = np.array(data)
    N = len(array)
    mu = sum(array)/N
    sigm = sqrt(np.sum((array-mu)**2)/N)
    return sigm

print(f"The standard deviation calculated using functions: {stdev([1, 6, 4, 10, 31, 5.1])}.")
print("The standard deviation calculated using np.std():", np.std([1, 6, 4, 10, 31, 5.1]),".")
```

```python
array = np.array([1, 6, 4, 10, 31, 5.1])
sigma = sqrt(np.sum((array-sum(array)/len(array))**2/len(array)))
print(f"The standar deviation calculated using a one-liner: {sigma}.")
```

### Q3.18 (session quiz)

```python
import numpy as np
#Seperating the table into two NumPy arrays:
data = [18.49, 2094.75,
35.48, 2320.87,
35.48, 1224.9,
11.34, 872.13,
7.12, 432.2,
9.82, 751,
23.66, 1410,
44, 2700,
66.7, 5292,
52.4, 2550,
55, 5253,
25.6, 2419,
26.19, 2158]

distances = np.array(data[0::2])
velocities = np.array(data[1::2])
print(f"List of distances: {distances}.", f"List of velocities: {velocities}.", sep='\n')

#Calculating a and b:
part_x = distances - np.sum(distances)/distances.size
part_y = velocities - np.sum(velocities)/velocities.size

b = np.sum(part_x*part_y)/np.sum(part_x**2)
a = np.sum(velocities)/velocities.size - b*np.sum(distances)/distances.size
print()
print(f"a = {a}, b = {b}") 
```

#### Using the range function for the x values:

```python
from math import exp
from matplotlib import pyplot as plt

#Values
x = range(-5, 6)
y = []
for p in x:
    func = (exp(p)+exp(-p))/2
    y.append(func)

#Plot
plt.plot(x, y, marker='o', color='r', label="Data")
plt.title(r"Catenary function", fontsize = 15)
plt.grid()
plt.ylabel(r"Height (m)")
plt.xlabel(r"x (m)")
plt.legend()
plt.show()
```

#### Using np.arange for the x values:

```python
import numpy as np
from math import exp
from matplotlib import pyplot as plt

#Values
x = np.arange(-5, 6)
y = []
for p in x:
    func = (exp(p)+exp(-p))/2
    y.append(func)

#Plot
plt.plot(x, y, marker='o', color='r', label="Data")
plt.title(r"Catenary function")
plt.grid()
plt.ylabel(r"Height (m)")
plt.xlabel(r"x (m)")
plt.legend()
plt.show()
```
