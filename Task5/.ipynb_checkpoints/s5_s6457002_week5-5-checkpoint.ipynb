### Q5.3 (coursework)

```python
import numpy as np
from scipy.constants import G
import matplotlib.pyplot as plt

#Seperating columns of rotcur.dat:
data = np.loadtxt('/Users/users/mbweeren/PROGNUM-repo/Task5/rotcur.dat')
radius = data[:, 0]
v = data[:, 1]

#Defining variables:
M_sun = 1.989e30
M_star = 2e10 * M_sun

r = radius[1:] * 3.086e19
vkepms = np.sqrt(G * M_star / r)
vkep = vkepms / 1e3

rkep = radius[1:]

#Creating and saving textfile keplerian_curve.txt:
keplerian_curve = np.column_stack((rkep, vkep))
np.savetxt('keplerian_curve.txt', keplerian_curve)

#Loading data:
kep_data = np.loadtxt('keplerian_curve.txt')
r_kep_load = kep_data[:, 0]
v_kep_load = kep_data[:, 1]

plt.figure(figsize=(9, 6))

plt.plot(radius, v, color='blue', linewidth=2, label='NGC 6946 observed rotation curve')
plt.plot(r_kep_load, v_kep_load, color='red', linewidth=2, label='Keplerian curve')

#Setting the limits for the curves:
plt.ylim(0, v_kep_load.max())
plt.xlim(0, radius.max())

#Plotting curve:
plt.xlabel('Radius (kpc)', fontsize=13)
plt.ylabel('Rotation velocity (km/s)', fontsize=13)
plt.title('NGC 6946 Galaxy Rotation Curve', fontsize=15)
plt.legend(fontsize=11)
plt.grid(True)

plt.tight_layout()
plt.savefig('rotation_curve.png')
plt.show()
```

### Q5.9 (session quiz)

```python
import imageio.v3 as imageio  # Using v2 for consistent behavior across environments
import matplotlib.pyplot as plt
import numpy as np

# imageio.imread reads the image file and converts it into a NumPy array.
# If the image is BW (Black and White), it will likely be a 2D array.
# If it is a JPG saved with RGB channels, we will handle that to ensure it is 2D.
img = imageio.imread('m101BW.jpg')

# Ensure the data is 2D (grayscale). If it has multiple channels, take the first one.
if len(img.shape) == 3:
    data = img[:, :, 0]
else:
    data = img

# The 'jet' colormap maps low intensities to blue and high intensities to red.
plt.figure(figsize=(8, 6))
plt.title("Original Image with 'jet' Colormap")
plt.imshow(data, cmap='jet')
plt.colorbar(label='Intensity Value')  # Adds a bar showing the color-to-value mapping
plt.show()

# Noise typically resides in the lower intensity values (the dark background).
# By setting 'levels' to start above the background noise floor, we only contour the galaxy.
# Here, we define levels starting from the 75th percentile to skip the noisy background.
noise_floor = np.percentile(data, 75)
max_intensity = np.max(data)
levels = np.linspace(noise_floor, max_intensity, 10) # 10 contours from noise floor to max

plt.figure(figsize=(8, 6))
plt.title("Contour Plot (Filtering Background Noise)")
# origin='image' ensures the contour orientation matches the imshow orientation
plt.contour(data, levels=levels, cmap='jet', origin='image')
plt.colorbar(label='Intensity Level')
plt.gca().invert_yaxis() # Invert Y-axis to match the image coordinate system
plt.show()

# To find the top 10% intensity, we calculate the 90th percentile of all pixel values.
threshold_10_percent = np.percentile(data, 90)

# Create a new image where pixels below the threshold are set to 0 (black).
# np.where(condition, value_if_true, value_if_false)
filtered_data = np.where(data >= threshold_10_percent, data, 0)

# Plot the filtered image
plt.figure(figsize=(8, 6))
plt.title("Top 10% Highest Intensity Features")
plt.imshow(filtered_data, cmap='jet')
plt.colorbar(label='Intensity Value')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# np.loadtxt reads the ASCII file into a 1D NumPy array.
# Make sure the file 'ngc6946.txt' is in the same directory as your script.
try:
    data_raw = np.loadtxt('/Users/users/mbweeren/PROGNUM-repo/Task5/ngc6946.txt')
except FileNotFoundError:
    print("Error: 'ngc6946.txt' not found. Please ensure the file is in the working directory.")
    # Creating dummy data for demonstration if file is missing (comment out in actual use)
    data_raw = np.random.normal(10, 2, 40 * 50 * 70) 

# The prompt specifies: 40 planes (z), 50 rows (y), and 70 columns (x).
# Reshaping converts the flat 1D array into a 3D "cube".
data_cube = data_raw.reshape((40, 50, 70))

# Calculate the middle indices for the x and y dimensions.
mid_y = 50 // 2  # Middle row (y-axis)
mid_x = 70 // 2  # Middle column (x-axis)

# A z-profile is a line of data along the z-axis (planes) at a fixed (y, x) point.
# Using ":" for the first dimension selects all 40 planes.
z_profile = data_cube[:, mid_y, mid_x]

# Using NumPy's built-in functions to analyze the extracted profile.
p_max = np.max(z_profile)
p_min = np.min(z_profile)
p_sum = np.sum(z_profile)
p_mean = np.mean(z_profile)
p_std = np.std(z_profile)

# Prepare a text string to display these statistics on the plot.
stats_text = (
    f"Max: {p_max:.2f}\n"
    f"Min: {p_min:.2f}\n"
    f"Sum: {p_sum:.2f}\n"
    f"Mean: {p_mean:.2f}\n"
    f"Std Dev: {p_std:.2f}"
)

plt.figure(figsize=(10, 6))

# Plot the z-profile (z-index vs intensity)
plt.plot(z_profile, marker='o', linestyle='-', color='teal', label='Z-Profile Data')

# plt.text(x_coord, y_coord, string, transform=...)
# Using transform=plt.gca().transAxes allows us to use relative coordinates (0 to 1).
# (0.05, 0.95) places the text box in the top-left corner of the plot area.
plt.text(0.05, 0.95, stats_text, transform=plt.gca().transAxes, 
         fontsize=10, verticalalignment='top', 
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))

# Formatting the plot
plt.title(f"Z-Profile at Coordinates (y={mid_y}, x={mid_x})")
plt.xlabel("Plane Index (Z-axis)")
plt.ylabel("Intensity / Value")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='upper right')

# Save and show the plot
plt.savefig('z_profile_ngc6946.png')
plt.show()

# Print confirmation
print(f"Profile extracted at (y={mid_y}, x={mid_x}).")
print(f"Statistics:\n{stats_text}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# We assume the data from 'ngc6946.txt' is a flat sequence of numbers.
# As specified, the cube has dimensions (40, 50, 70).
try:
    data_raw = np.loadtxt('/Users/users/mbweeren/PROGNUM-repo/Task5/ngc6946.txt')
    # Reshape the 1D array into the 3D cube (Planes, Rows, Columns)
    data_cube = data_raw.reshape((40, 50, 70))
except FileNotFoundError:
    print("Error: 'ngc6946.txt' file not found.")
    # Fallback to dummy data if the file is missing for demonstration purposes
    data_cube = np.random.normal(50, 15, (40, 50, 70))

# In a (z, y, x) structure, the first plane is index 0 of the z-dimension.
# This represents the image at the first 'depth' level.
first_plane = data_cube[0, :, :]

# We need the mean and standard deviation of all pixels in this specific plane.
p_mean = np.mean(first_plane)
p_std = np.std(first_plane)

# Define the 3 specific contour levels requested:
# - Level 1: The average intensity (mean)
# - Level 2: One standard deviation above the mean
# - Level 3: Two standard deviations above the mean (highly intense regions)
contour_levels = [p_mean, p_mean + p_std, p_mean + 2 * p_std]

plt.figure(figsize=(10, 8))

# Using plt.imshow() to display the 2D array as an image.
# argument origin='lower' ensures the (0,0) pixel is at the bottom-left corner.
# This is common in astronomical or scientific mapping to match coordinate systems.
img = plt.imshow(first_plane, origin='lower', cmap='plasma')

# Add a color bar to correlate the colors with the actual data values.
plt.colorbar(img, label='Intensity Value')

# plt.contour() draws lines connecting points of equal value.
# origin='lower' must be set here as well to align the contours with the image.
# We use 'white' for the lines so they stand out against the 'plasma' colormap.
contours = plt.contour(first_plane, levels=contour_levels, 
                       origin='lower', colors='white', linewidths=1.5)


plt.title(f"First Plane of Data Cube (z=0)\nOrigin: Lower Left, Contours: [mean, mean+std, mean+2*std]")
plt.xlabel("X-Axis (70 Columns)")
plt.ylabel("Y-Axis (50 Rows)")

plt.show()
```

### Q5.12 (session quiz)

```python
import scipy.constants

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
    caH_0 = 3968.5      # Calcium H line in rest frame
    c = scipy.constants.c/1000
    
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
        
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    
    def velocity(self):
        z=self.redshift()
        v = z*self.c
        return v

Milkyway = Galaxy("Milkyway", calciumHline = 3970.0) 
print(f"The velocity of the galaxy is: {Milkyway.velocity()} km/s")
```

### Q5.14 (coursework)

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Ellipse

#Copied code
class AstroEllipse(Ellipse):
    def __init__(self, xc, yc, semi_major, semi_minor, position_angle, **args):
      # We use the (sort of) astronomical angle where 'angle' defines)
      # the position angle of the major axis of a ring projected onto the sky
      # (i.e. an ellipse). This is an angle taken in anti-clockwise direction between
      # the north direction on the sky and the major axis of the receding half of the
      # galaxy (Rots 1975, astron, astrophys 45, 43).
      # Here we assume (!) north is in the positive Y direction.
      # If not, you need to correct the angle for the rotation of the image.
      # Parameter 'position_angle' enters in degrees
      assert(semi_major >= semi_minor)  # Prerequisite to make angle unambiguous
      XY = (xc, yc)                  # MPL requires center as a tuple
      angle_mpl = position_angle     # Angle for MatPlotLib's ellipse patch
      height = 2.0*semi_major        # height: is along vertical axis for an unrotated ellipse
      width  = 2.0*semi_minor        # width:  is along horizontal axis for an unrotated ellipse
      super().__init__(XY, width, height, angle=angle_mpl, **args)
    
    #Letting the arguments of the previous function be variables:
      self.xc = xc
      self.yc = yc
      self.semi_major = semi_major
      self.semi_minor = semi_minor
      self.position_angle = position_angle

    #Function calculating the area of the ellipse:
    def area(self):
        return np.pi * self.semi_major * self.semi_minor
    
    #Function calculating the perimiter of the ellipse:
    def perimeter(self):
        a = self.semi_major
        b = self.semi_minor
        h = ((a - b) ** 2) / ((a + b) ** 2)
        return np.pi * (a + b) * (1 + (3*h)/(10 + np.sqrt(4 - 3*h)))
    #Method _str_:
    def __str__(self):
        return (f"AstroEllipse(center=({self.xc}, {self.yc}), "
                f"semi_major={self.semi_major}, semi_minor={self.semi_minor}, "
                f"angle={self.position_angle})")

#Creating class MyCircle
class MyCircle(AstroEllipse):
    def __init__(self, xc, yc, radius, **args):
        super().__init__(xc, yc, semi_major=radius, semi_minor=radius, 
                         position_angle=0, **args)

ellipse = AstroEllipse(0, 0, 3, 2, 45)
print("Example of an ellipse:", ellipse)
print(f"The area of the ellipse is: {ellipse.area()}")
print(f"The Perimeter of the ellipse is: {ellipse.perimeter()}")

#Creating plot:
fig, ax = plt.subplots(figsize=(8, 8))

np.random.seed(42)
n_ellipses = 5
n_circles  = 5

for _ in range(n_ellipses):
    xc    = np.random.uniform(-5, 5)
    yc    = np.random.uniform(-5, 5)
    smaj  = np.random.uniform(0.5, 2.0)
    smin  = np.random.uniform(0.1, smaj) 
    angle = np.random.uniform(0, 180)
    color = np.random.rand(3)
    ell = AstroEllipse(xc, yc, smaj, smin, angle,
                       color=color, alpha=0.5)
    ax.add_patch(ell)

for _ in range(n_circles):
    xc     = np.random.uniform(-5, 5)
    yc     = np.random.uniform(-5, 5)
    radius = np.random.uniform(0.2, 1.0)
    color  = np.random.rand(3)
    circ = MyCircle(xc, yc, radius, color=color, alpha=0.5)
    ax.add_patch(circ)

ax.set_xlim(-8, 8)
ax.set_ylim(-4, 8)
ax.set_aspect('equal')
ax.set_title("Some circles and ellipses")
plt.show()
```

### Q5.15 (coursework)

```python
from scipy.special import jv
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 10, 1000)
a_values = [1, 2, 3, 4]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Bessel Functions of the first kind of real order a", fontsize=14)

for ax, a in zip(axes.flatten(), a_values):
    ax.plot(x, jv(a, x), label=f'J_{a}(x)')
    ax.set_title(f'a = {a}')
    ax.set_xlabel('x')
    ax.set_ylabel('J(x)')
    ax.legend()
    ax.grid(True)
    
#Plotting and saving figure:
plt.tight_layout()
plt.savefig('bessel.png', transparent=True)
plt.show()
```
